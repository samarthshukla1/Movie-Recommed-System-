# import Tfidf vectorization 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfy = TfidfVectorizer(min_df=3, max_features = None, strip_accents="unicode",
                      analyzer="word", token_pattern=r"\w{1,}", ngram_range=(1,3), stop_words = "english")
tfv_matrix = tfy.fit_transform(merged_dataframe_cleaned["overview"].values.astype(str))
print(tfv_matrix)
print(tfv_matrix.shape)
```

# Using sigmoid kernel function on tfidf vectorization

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig_ker = sigmoid_kernel(tfv_matrix, tfv_matrix)
sig_ker
```

# Defining function for recomendation

```python
def recommend_movie(title, sig=sig_ker):
    index = indices[title]
    sig_scores = list(enumerate(sig[index]))
    sig_scores = sorted(sig_scores, key = lambda x: x[1],reverse=True)
    movie_indices = [i[0] for i in sig_scores]
    sig_scores = sig_scores[1:11]
    return merged_dataframe_cleaned["original_title"].iloc[movie_indices]
```

# Movie Recommedation on basis of cast crew


```python
# Parse the stringified features into their corresponding python objects
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    merged_dataframe[feature] = merged_dataframe[feature].apply(literal_eval)
```

```python
#Get the director's name from the crew feature. If director is not listed, return NaN
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
#Returns the list top 3 elements or entire list; whichever is more.
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
  #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
#Define new director, cast, genres and keywords features that are in a suitable form.
merged_dataframe['director'] = merged_dataframe['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    merged_dataframe[feature] = merged_dataframe[feature].apply(get_list)
```

```python
# Print the new features of the first 5 films
merged_dataframe[['title_x', 'cast', 'director', 'keywords', 'genres']].head(5)
```

```python
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Apply clean_data function to your features.
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    merged_dataframe[feature] = merged_dataframe[feature].apply(clean_data)
```

# Creating metadata

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
merged_dataframe['soup'] = merged_dataframe.apply(create_soup, axis=1)
```

```python
# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(merged_dataframe['soup'])
```

```python
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Reset index of our main DataFrame and construct reverse mapping as before
merged_dataframe = merged_dataframe.reset_index()
indices = pd.Series(merged_dataframe.index, index=merged_dataframe['title_x'])
```
